```python
# We're going to make it a time series
#============================================
series_years_NGCBMU =df0_NGCBMU_years.to_period('Y')
series_years_NGCBMUsstt = series_years_NGCBMU.to_timestamp('Y')
series_years_NGCBMUsstt.describe()
```

```python
# Detailed descriptive
#=================================================
import pandas as pd
from scipy.stats import describe, variation, shapiro

# Calculate mean, median, quartiles, variance, standard deviation, Pearson's coefficient of variation, maximum and minimum.
print('Mean:', series_years_NGCBMUsstt.mean())
print('Median:', series_years_NGCBMUsstt.median())
print('Quartiles:', series_years_NGCBMUsstt.quantile([0.25, 0.5, 0.75]))
print('Variance:', series_years_NGCBMUsstt.var())
print('Standard deviation:', series_years_NGCBMUsstt.std())
print('Pearsons coefficient of variation:', variation(series_years_NGCBMUsstt)*100, '%')
print('Maximum:', series_years_NGCBMUsstt.max())
print('Minimmum:', series_years_NGCBMUsstt.min())

# Number of observartions
print('Number of observations:', len(series_years_NGCBMUsstt))

# Shapiro-Wilk normality test
stat, p = shapiro(series_years_NGCBMUsstt)
alpha = 0.05
if p > alpha:
  print('The time series appears to be normal (p = ', p, ')')
else:
  print('The time series does not appear to be normal (p = ', p, ')')
```

```python
import numpy as np
from scipy.stats import t

def grubbs_test(serie_years_NGCBMUsstt, alpha):
    """
    Performs the Grubbs test to detect outliers in a one-dimensional data series.    
    The null hypothesis is that there are no outliers in the data.    
    Returns a tuple with the calculated critical value, the observed value, 
    and True if the null hypothesis is rejected (there is at least one outlier) or False otherwise.   
    """
    n = len(series_years_NGCBMUsstt)
    mean = np.mean(series_years_NGCBMUsstt)
    std = np.std(series_years_NGCBMUsstt, ddof=1)
    t_crit = t.ppf(1 - alpha / (2*n), n-2)
    g_crit = (n-1) / np.sqrt(n) * np.sqrt(t_crit**2 / (n-2 + t_crit**2))
    g_obs = np.max(np.abs(series_years_NGCBMUsstt - mean)) / std
    return g_crit, g_obs, g_obs > g_crit

# Ejemplo de uso:
alpha = 0.05
g_crit, g_obs, is_outlier = grubbs_test(series_years_NGCBMUsstt, alpha)
print("Crticial value: ", g_crit)
print("Observed value: ", g_obs)
if is_outlier:
    print("There are outliers in the data.")
else:
    print("There are no outliers in the data.")
```
